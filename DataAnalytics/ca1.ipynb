```python
# a) 
import numpy as np
data = np.array([20, 21, np.nan, 22, 210, 19, 20], dtype=float)
median = np.nanmedian(data)
data[np.isnan(data)] = median
print("Median:", median)
print("After replacing missing value:", data)


#b)
sorted_data = np.sort(data)
Q1 = np.percentile(sorted_data, 25)
Q3 = np.percentile(sorted_data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("\nQ1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Bounds:", (lower_bound, upper_bound))
outliers = sorted_data[(sorted_data < lower_bound) | (sorted_data > upper_bound)]
print("Outliers:", outliers)
```

```python
#2)
import numpy as np
data = np.array([10, 11, 12, 13, 150])
mean = np.mean(data)
std = np.std(data)
z_score = (150 - mean) / std
print("Mean:", mean)
print("Standard Deviation:", std)
print("Z-score of 150:", z_score)
if abs(z_score) >= 2:
    print("Potential anomaly")
else:
    print("Normal value")
```

```python
#3)
speeds = [20, 30, 40, 50, 60, 70, 80]
x = 50
xmin = min(speeds)
xmax = max(speeds)
normalized = (x - xmin) / (xmax - xmin)
print("Normalized value:", normalized)
```
