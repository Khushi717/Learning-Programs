```python
import pandas as pd
import numpy as np
data={
    "population":[np.nan,45,56,90],
    "capital":[np.nan,"Berlin","Paris","Ghaziabad"]
}

df=pd.DataFrame(data,index=["India","Austrailia","America","Nepal"])
print(df)
```

```python
df.fillna({"population":df["population"].mean()},inplace=True)
```

```python
df.fillna({"capital":df["population"].mode()[0]},inplace=True)
```
